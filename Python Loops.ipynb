# Break, Continue, pass

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(1,11):
    if i == 4:
        break
    print(i)
```

```python
for i in range(1,11):
    if i == 4:
        continue
    print(i)
```

```python
for i in range(1,11):
    if i == 4:
        pass
    print(i)
```

```python
for i in range(1,11):
    pass
```
