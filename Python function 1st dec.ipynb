# Keyword Argument

```python
def person(name,age):
    print(name)
    print(age+1)
    
person(age=23, name='Abid')
```

```python
def person(name,age1):
    print(name)
    print(age1+1)

person(age1=23, name='nit')   
```

```python
def person(name,age,city):
    print(name)
    print(age)
    print(city)

person(age=23, name='Abid', city='Siwan')
```

```python
def person(name,age=18):
    print(name)
    print(age)

person('nit', 24)
```
